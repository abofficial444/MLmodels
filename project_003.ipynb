```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
estate = pd.DataFrame ([("Lekki",2600, 550000),
                       ("Lekki",3000, 565000 ),
                       ("Lekki", 3200, 610000),
                       ("Lekki", 3600, 680000),
                       ("Lekki", 4000, 725000),
                       ("Banana island",2600, 585000),
                       ("Banana island", 2800, 615000),
                       ("Banana island", 3300, 650000),
                       ("Banana island", 3600, 710000),
                       ("Ajah", 2600, 575000),
                       ("Ajah", 2900, 600000),
                       ("Ajah", 3100, 620000),
                       ("Ajah", 3600, 695000),],
                        columns = ["CITY", " SIZE", "PRICE"],)
estate
```

```python
encoding = pd.get_dummies(estate.CITY)
encoding
```

```python
merged = pd.concat([estate, encoding], axis ="columns")
merged
```

```python
drop_town = merged.drop(["CITY", "Ajah"],  axis = "columns")
drop_town 
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
```

```python
X = drop_town.drop("PRICE", axis ="columns")
X
```

```python
Y = drop_town.PRICE
Y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size=0.3)
print(Y_train.shape, Y_test.shape)
```

```python
model.fit(X_train,Y_train)
```

```python
model.predict(X_test)
```

```python
Y_test
Y_test_reshaped = np.array(Y_test).reshape((1,4))
Y_test_reshaped
```

```python
model.score(X_test, Y_test)
```
